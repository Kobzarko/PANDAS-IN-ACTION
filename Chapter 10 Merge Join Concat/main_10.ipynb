### Missing Values in Combined DataFrames

```python
sports_champions_A = pd.DataFrame(
data = [
["New England Patriots", "Houston Astros"],
["Philadelphia Eagles", "Boston Red Sox"]
],
columns = ["Football", "Baseball"],
index = [2017, 2018]
)
```

```python
sports_champions_B = pd.DataFrame(
data = [
["New England Patriots", "St. Louis Blues"],
["Kansas City Chiefs", "Tampa Bay Lightning"]
],
columns = ["Football", "Hockey"],
index = [2019, 2020]
)
```

**Try to concatenate and observe NaN values at the intersection of rows and columns that are not common to the source datasets**

```python
pd.concat(objs = [sports_champions_A, sports_champions_B])
```

**By default, pandas performs the concatenation of horizontal lines. However, sometimes it is preferable to concatenate vertically.**

```python
sports_champions_C = pd.DataFrame(
data = [
["Pittsburgh Penguins", "Golden State Warriors"],
["Washington Capitals", "Golden State Warriors"]
],
columns = ["Hockey", "Basketball"],
index = [2017, 2018]
)
sports_champions_C
```

**By horizontal**

```python
pd.concat(objs = [sports_champions_A, sports_champions_C])
```

**By vertical**

```python
pd.concat(objs = [sports_champions_A, sports_champions_C], axis = 'columns')
```

```python
week1  = pd.read_csv('/Users/ypushiev/Learning/PANDAS IN ACTION/Chapter 10 Merge Join Concat/Data/Restaurant/week_1_sales.csv')
week2  = pd.read_csv('/Users/ypushiev/Learning/PANDAS IN ACTION/Chapter 10 Merge Join Concat/Data/Restaurant/week_2_sales.csv')
customers = pd.read_csv('/Users/ypushiev/Learning/PANDAS IN ACTION/Chapter 10 Merge Join Concat/Data/Restaurant/customers.csv', index_col='ID')
food  = pd.read_csv('/Users/ypushiev/Learning/PANDAS IN ACTION/Chapter 10 Merge Join Concat/Data/Restaurant/foods.csv', index_col='Food ID')
```

#### Find customers who have been to the restaurant at each review weeks.
