### Aggregate Operations

**The sum method is applied to a GroupBy object and calculates sums by sectors for three numeric columns (Revenue, Profits, and Employees (Number of employees)) in the Fortune dataset**

```python
sectors.sum().head(10)
```

**Check results in Aerospace & Defense**

```python
sectors.get_group("Aerospace & Defense").head()
```

```python
sectors.get_group("Aerospace & Defense").loc[:,"Revenues"].sum()
```

**call the mean method, which calculates the average values of the Revenue, Profits, and Employees columns for each sector.**

```python
sectors[['Revenues','Profits','Employees']].mean().head()
```

**The max method returns the maximum value in the specified column.**

```python
sectors["Profits"].max().head()
```

**The min method returns the minimum value in the specified column.**

```python
sectors["Employees"].min().head()
```

**The agg method applies several aggregate operations to different columns and accepts a dictionary as an argument.**

```python
aggregations = {
"Revenues": "min",
"Profits": "max",
"Employees": "mean"
}

sectors.agg(aggregations).head()
```

### Applying a Custom Operation to all Groups

**The nlargest DataFrame object method retrieves records with the highest value
in the specified column.**

```python
df_fortune.nlargest(n = 5, columns = "Profits")
```

**Create a custom function that returns the largest value in the Revenues column**

```python
def get_largest_row(df):
    return df.nlargest(1, "Revenues")
```

```python
sectors.apply(get_largest_row).head()
```
