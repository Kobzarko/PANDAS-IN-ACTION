```python
df_sales.pivot_table(index="Date", values=['Revenue','Expenses'], aggfunc='mean')
```

**Create pivot table using columns**

```python
print(df_sales.pivot_table(index="Date",columns='Name', values='Revenue', aggfunc='sum'))
```

**Replace NaN with fill_value=0**

```python
print(df_sales.pivot_table(index="Date",columns='Name', values='Revenue', aggfunc='sum', fill_value=0))
```

**Summarize outcomes with margins=True**

```python
print(df_sales.pivot_table(index="Date",columns='Name', values='Revenue', aggfunc='sum', fill_value=0, margins=True))
```

**Summarize outcomes with margins=Total**

```python
print(df_sales.pivot_table(index="Date",columns='Name', values='Revenue', aggfunc='sum', fill_value=0, margins='Total'))
```

### Additional Options for Pivot Tables

**Count employee sales**

```python
print(df_sales.pivot_table(index="Date",columns='Name', values='Revenue', aggfunc='count', fill_value=0))
```

**Use two operations**

```python
print(df_sales.pivot_table(index="Date",columns='Name', values='Revenue', aggfunc=['count','sum'], fill_value=0))
```

**Different operations in different columns**

```python
print(df_sales.pivot_table(index="Date",columns='Name', values=['Expenses','Revenue'], aggfunc={'Expenses': 'max', 'Revenue': 'min'}, fill_value=0))
```

```python
df_sales.pivot_table(index="Date",columns='Name', values=['Expenses','Revenue'], aggfunc={'Expenses': 'max', 'Revenue': 'min'}, fill_value=0)
```

**Use multiIndex**

```python
print(df_sales.pivot_table(index=["Name", "Date"], values="Revenue", aggfunc="sum").head(10))
```

**Replace levels**

```python
print(df_sales.pivot_table(index=["Date", "Name"], values="Revenue", aggfunc="sum").head(10))
```

### Stacking and Unstacking Index Levels

```python
by_name_and_date = df_sales.pivot_table(index="Name", columns = 'Date', values="Revenue", aggfunc="sum", fill_value=0)
```

```python
print(by_name_and_date.head())
```

**Change axis for index using Stack()**

```python
print(by_name_and_date.stack().head(10))
```

**Change axis for index using Unstack()**

```python
print(by_name_and_date.T)
```

```python
sales_by_customer = df_sales.pivot_table(
index = ["Customer", "Name"],
values = "Revenue",
aggfunc = "sum",
fill_value=0
)
```

```python
print(sales_by_customer.head(10))
```

```python
print(sales_by_customer.unstack().head(10))
```
