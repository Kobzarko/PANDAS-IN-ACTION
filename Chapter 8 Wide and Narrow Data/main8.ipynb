```python
print(df_sales.pivot_table(index="Date",columns='Name', values='Revenue', aggfunc='sum', fill_value=0, margins='Total'))
```

### Additional Options for Pivot Tables

**Count employee sales**

```python
print(df_sales.pivot_table(index="Date",columns='Name', values='Revenue', aggfunc='count', fill_value=0))
```

**Use two operations**

```python
print(df_sales.pivot_table(index="Date",columns='Name', values='Revenue', aggfunc=['count','sum'], fill_value=0))
```

**Different operations in different columns**

```python
print(df_sales.pivot_table(index="Date",columns='Name', values=['Expenses','Revenue'], aggfunc={'Expenses': 'max', 'Revenue': 'min'}, fill_value=0))
```

```python
df_sales.pivot_table(index="Date",columns='Name', values=['Expenses','Revenue'], aggfunc={'Expenses': 'max', 'Revenue': 'min'}, fill_value=0)
```

**Use multiIndex**

```python
print(df_sales.pivot_table(index=["Name", "Date"], values="Revenue", aggfunc="sum").head(10))
```

**Replace levels**

```python
print(df_sales.pivot_table(index=["Date", "Name"], values="Revenue", aggfunc="sum").head(10))
```

### Stacking and Unstacking Index Levels

```python
by_name_and_date = df_sales.pivot_table(index="Name", columns = 'Date', values="Revenue", aggfunc="sum", fill_value=0)
```

```python
print(by_name_and_date.head())
```

**Change axis for index using Stack()**

```python
print(by_name_and_date.stack().head(10))
```

**Change axis for index using Unstack()**

```python
sales_by_customer = df_sales.pivot_table(
index = ["Customer", "Name"],
values = "Revenue",
aggfunc = "sum",
fill_value=0
)
```

```python
print(sales_by_customer.head(10))
```

```python
print(sales_by_customer.unstack().head(10))
```

```python
cars_path = '/Users/ypushiev/Learning/PANDAS IN ACTION/Chapter 8 Wide and Narrow Data/Data8/used_cars.csv'
min_wage_path = '/Users/ypushiev/Learning/PANDAS IN ACTION/Chapter 8 Wide and Narrow Data/Data8/minimum_wage.csv'
```

```python
df_cars = pd.read_csv(cars_path)
df_min_wage = pd.read_csv(min_wage_path)
```

```python
df_cars.nunique()
```

```python
df_cars.head()
```

**Aggregate prices of cars. Group results by type fuel along the line axis.**

```python
df_cars.pivot_table(index='Fuel', values='Price', aggfunc='sum', fill_value=0)
```

**Aggregate the number of cars in the cars. Group the results according to the driver on the line axis and gearbox type on the column axis.**

```python
df_cars.pivot_table(index='Manufacturer', columns = 'Transmission', values='Price', aggfunc='count', fill_value=0, margins=True )
```

**Aggregate the average price of cars. Group results by year and the fuel type on the line axis, and the gearbox type on the column axis.**

```python
df_cars.pivot_table(index=['Year','Fuel'], columns='Transmission', values='Price', aggfunc='mean', fill_value=0)
```

**DataFrame object from previous task transfers level Transmission from the column axis to the line axis.**

```python
report = df_cars.pivot_table(
values = "Price",
index = ["Year", "Fuel"],
columns = ["Transmission"],
aggfunc = "mean"
)
```

```python
print(report.stack())
```
